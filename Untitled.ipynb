```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Read the CSV file into a Dataframe 
mouse_metadata = pd.read_csv('Mouse_metadata.csv')
study_results = pd.read_csv('study_results.csv')

#display both csv documents
print(mouse_metadata)
print(study_results)
```

```python
#Merge both CSV documents into one single DataFrame 
merged_data = pd.merge(mouse_metadata, study_results, on = 'Mouse ID', how = 'inner')
```

```python
#Calculate the number of unique mice IDs in the merged data 
unique_mice_count = merged_data['Mouse ID'].nunique()

#Check for duplicate Mouse ID's with duplicate Timepoints 
duplicate_mouse_ids = merged_data[merged_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep =False)]

#Display the data associated with Mouse ID's with duplicate time points
if not duplicate_mouse_ids.empty:
    print("Data associated with Mouse IDs with duplicate time points.")
    display(duplicate_mouse_ids)
else:
    print("No Mouse ID's with duplicate time points found.")

#Create a cleaned Dataframe without duplicate Time Points 
cleaned_data = merged_data.drop_duplicates(subset = ['Mouse ID', 'Timepoint'], keep = False)

print(cleaned_data)
```
